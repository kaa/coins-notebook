```python
from keras.preprocessing.image import ImageDataGenerator
from keras import backend as K
from keras import optimizers
from keras.models import Model
from keras.layers import Dense
from deepcash import DeepCash
import densenet
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.filters
```

```python
# dimensions of our images.
img_width, img_height = 80, 80

train_data_dir = 'dataset/classified/train'
validation_data_dir = 'dataset/classified/test'
nb_train_samples = 864
nb_validation_samples = 442
epochs = 300
batch_size = 16
input_shape = (img_width, img_height, 3)

model = DeepCash(input_shape, classes = 8)
#model = densenet.DenseNet(input_shape, classes = 8, activation = "softmax")
#base_model = densenet.DenseNet(input_shape=input_shape, weights="imagenet", include_top=False)
#x = base_model.output
#x = Dense(8, activation="softmax")(x)
#model = Model(inputs=base_model.input, outputs=x)

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.Adam(decay=1e-4),
              metrics=['accuracy'])

def blur(img):
    img = img*(1/255.0)
    img = skimage.filters.gaussian(img, sigma=0.5)
    return img

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
#    rescale=1./255,
    shear_range=0.1,
    zoom_range=0.1,
    rotation_range=90,
    horizontal_flip=False,
    preprocessing_function=blur)

# this is the augmentation configuration we will use for testing:
# only rescaling
test_datagen = ImageDataGenerator(
#    rescale=1./255,
    preprocessing_function=blur)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size)

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size)
```

```python
model.summary()
```

```python
history = model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)

model.save("coins_model.h5")    
```

```python
# summarize history for accuracy
for key in ["acc","loss"]:
    plt.plot(history.history[key])
    plt.plot(history.history['val_'+key])

    plt.title('model '+key)
plt.ylabel(key)
plt.xlabel('epoch')
plt.legend(['acc', 'val_acc','loss','val_loss'], loc='upper left')
plt.show()
```
